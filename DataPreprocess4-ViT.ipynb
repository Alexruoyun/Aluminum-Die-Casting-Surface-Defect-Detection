```python
# Function for retrieving images by sample index and picture index, then resize, save to folder
def img_vit(src_dir, idx, IMG_SIZE, tar_dir):
    for i in idx:
        img_path = f"{src_dir}/{str(i).rjust(4, '0')}_p.PNG"  # Construct image path
        
        img = cv2.imread(img_path)  # Read image
        if img is None:
            print(f"Warning: Image {img_path} not found, skipping.")
            continue  # Skip if image is missing
        
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE), interpolation=cv2.INTER_AREA)  # Resize

        save_img_path = f"{tar_dir}/{str(i).rjust(4, '0')}_pp.PNG"  # Save path
        matplotlib.image.imsave(save_img_path, img)  # Save resized image
```

# Data Using ViT_ Data Folders

```python
# Function to load images from ViT data folders
def dt(src_path):
    temp_lst = []
    img_lst = os.listdir(src_path)
    for i in img_lst:
        f_path = src_path + '/' + i
        img = cv2.imread(f_path)
        if img.shape != (224, 224, 3): 
            print(f"Warning: Image size mismatch {f_path}")  # Improved warning message
            continue
        temp_lst.append(img)
    x = np.stack(temp_lst, axis=0)
    x = np.transpose(x, (0, 3, 2, 1))
    return x
```

```python
# Load training data
X_train_crack = dt("C:/Users/101194208/Desktop/Die-Cast Ensemble/Data/ViT_train/1")
X_train_good = dt("C:/Users/101194208/Desktop/Die-Cast Ensemble/Data/ViT_train/0")
X_train_cf = dt("C:/Users/101194208/Desktop/Die-Cast Ensemble/Data/ViT_train/2")

X_train = np.concatenate((X_train_crack, X_train_good, X_train_cf), axis=0)
y_train = [1] * X_train_crack.shape[0] + [0] * X_train_good.shape[0] + [2] * X_train_cf.shape[0]  # Updated labels
y_train = np.ravel(y_train)
```

```python
# Load validation data
X_val_crack = dt("C:/Users/101194208/Desktop/Die-Cast Ensemble/Data/ViT_val/1")
X_val_good = dt("C:/Users/101194208/Desktop/Die-Cast Ensemble/Data/ViT_val/0")
X_val_cf = dt("C:/Users/101194208/Desktop/Die-Cast Ensemble/Data/ViT_val/2")

X_val = np.concatenate((X_val_crack, X_val_good, X_val_cf), axis=0)
y_val = [1] * X_val_crack.shape[0] + [0] * X_val_good.shape[0] + [2] * X_val_cf.shape[0]  # Updated labels
y_val = np.ravel(y_val)
```

```python
# Load test data
X_test_crack = dt("C:/Users/101194208/Desktop/Die-Cast Ensemble/Data/ViT_test/1")
X_test_good = dt("C:/Users/101194208/Desktop/Die-Cast Ensemble/Data/ViT_test/0")
X_test_cf = dt("C:/Users/101194208/Desktop/Die-Cast Ensemble/Data/ViT_test/2")

X_test = np.concatenate((X_test_crack, X_test_good, X_test_cf), axis=0)
y_test = [1] * X_test_crack.shape[0] + [0] * X_test_good.shape[0] + [2] * X_test_cf.shape[0]  # Updated labels
y_test = np.ravel(y_test)
```

# Oversampling

```python
from imblearn.over_sampling import RandomOverSampler

# Reshape X_train to 2D for oversampling
n_samples, dim1, dim2, dim3 = X_train.shape
X_train_reshaped = X_train.reshape(n_samples, -1)  # Flatten the image features

# Apply Random Oversampling for multi-class balancing
ros = RandomOverSampler(sampling_strategy='auto', random_state=42)  # Balances all classes
X_train_resampled, y_train_resampled = ros.fit_resample(X_train_reshaped, y_train)

# Reshape X_train_resampled back to 4D
n_samples_resampled = X_train_resampled.shape[0]
X_train_resampled_4d = X_train_resampled.reshape(n_samples_resampled, dim1, dim2, dim3)

# Print shape verification
print(f'Original shape: {X_train.shape}')
print(f'Resampled shape: {X_train_resampled_4d.shape}')

# Check new class distribution after oversampling
unique, counts = np.unique(y_train_resampled, return_counts=True)
print('New class distribution after resampling:', dict(zip(unique, counts)))
```

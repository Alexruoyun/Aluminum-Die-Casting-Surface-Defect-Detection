```python
import csv
```

```python
def read_csv_to_lists(filename):
    true = []
    pred = []
    with open(filename, 'r') as csvfile:
        csv_reader = csv.reader(csvfile)
        for row in csv_reader:
            if row:
                true.append(int(row[0]))
                pred.append(int(row[1]))
    return true, pred
```

# Metrics

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
accuracy = accuracy_score(true_cnn, hard_voting_preds)
print(f"Accuracy: {accuracy}")
```

```python
# Calculate precision for each class without averaging
precision_per_class = precision_score(true_cnn, hard_voting_preds, average=None)
print("Precision per class:", precision_per_class)
```

```python
# Calculate macro-average precision (average of precision scores for each class)
macro_precision = precision_score(true_cnn, hard_voting_preds, average='macro')
print("Macro-average precision:", macro_precision)
```

```python
# Calculate recall for each class
per_class_recall = recall_score(true_cnn, hard_voting_preds, average=None)
print(f"Recall per class: {per_class_recall}")
```

```python
# Calculate macro-averaged recall
micro_recall = recall_score(true_cnn, hard_voting_preds, average='macro')
print(f"Macro-averaged recall: {micro_recall}")
```

```python
import os # OS module in Python provides functions for interacting with the operating system.
from PIL import Image # Open image
from rembg import remove # Image background removing
from PIL import ImageChops # Image crop
```

```python
# Function for image cropping
def trim(im):
    bg = Image.new(im.mode, im.size, im.getpixel((0,0)))
    diff = ImageChops.difference(im, bg)
    diff = ImageChops.add(diff, diff, 2.0, -100)
    #Bounding box given as a 4-tuple defining the left, upper, right, and lower pixel coordinates.
    #If the image is completely empty, this method returns None.
    bbox = diff.getbbox()
    if bbox:
        return im.crop(bbox)
```

## Crack

```python
# List file in directory
image_lst = os.listdir('C:/Users/101194208/Desktop/die_casting_ensemble/Data/Crack')

# For loop: read each image, remove background, crop, reduce resolution and save to folder
for i in image_lst:
    input_path = 'C:/Users/101194208/Desktop/die_casting_ensemble/Data/Crack/' + i
    output_path = 'C:/Users/101194208/Desktop/die_casting_ensemble/Data/Crack_rmbg_crop_rdrs/' + i[:4] + '_p.PNG' # _p processed once
    input = Image.open(input_path)
    x = remove(input) # Remove background
    output = trim(x) # Crop image
    output.save(output_path, quality = 1) # Save image and reduce resolution
```

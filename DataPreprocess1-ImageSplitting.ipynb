## Split the index into train, validation, and test

```python
import random # Import the random library
```

```python
# Parameters
num_crack = 894 # Number of samples; 
num_good = 2022
num_cf = 431
```

####  Crack indices

```python
crack_idx = list(range(1, num_crack+1))
train_idx_crack = random.sample(crack_idx, round(num_crack*0.8)) # 715 (80%) sample indices for training
```

```python
# Use a list comprehension to compute the difference while maintaining the original order
rest_idx = [item for item in crack_idx if item not in train_idx_crack]
# 90 (10%) sample indices for valid
val_idx_crack = [rest_idx[i] for i in random.sample(range(0, len(rest_idx)), round(len(rest_idx)*0.5))]
```

```python
test_idx_crack = [item for item in rest_idx if item not in val_idx_crack] # 89 (10%) sample indices for test
```

```python
print('train:', len(train_idx_crack), ', validation:', len(val_idx_crack), ', test:', len(test_idx_crack))
```

#### Good indices

```python
good_idx = list(range(1, num_good+1))
train_idx_good = random.sample(good_idx, round(num_good*0.8)) # 1618 (80%) sample indices for training
```

```python
rest_idx = [item for item in good_idx if item not in train_idx_good]
# 202 (10%) sample indices for valid
val_idx_good = [rest_idx[i] for i in random.sample(range(0, len(rest_idx)), round(len(rest_idx)*0.5))]
```

```python
test_idx_good = [item for item in rest_idx if item not in val_idx_good] # 202 (10%) sample indices for test
```

```python
print('train:', len(train_idx_good), ', validation:', len(val_idx_good), ', test:', len(test_idx_good))
```

#### CF indices

```python
cf_idx = list(range(1, num_cf+1))
train_idx_cf = random.sample(cf_idx, round(num_cf*0.8))  # 80% sample indices for training
```

```python
rest_idx = [item for item in cf_idx if item not in train_idx_cf]
val_idx_cf = [rest_idx[i] for i in random.sample(range(0, len(rest_idx)), round(len(rest_idx)*0.5))]  # 10% sample indices for validation
```

```python
test_idx_cf = [item for item in rest_idx if item not in val_idx_cf]  # 10% sample indices for test
```

```python
print('train:', len(train_idx_cf), ', validation:', len(val_idx_cf), ', test:', len(test_idx_cf))
```
